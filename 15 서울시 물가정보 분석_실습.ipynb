# 삼겹살 가격 분석

## 삼겹살 데이터
* 특정 문자열이 포함되어있는지 확인 : .str.contains(문자열)

```python
# 2021-06 데이터 이용
df_sam=df[ (df['품목 이름'].str.contains('삼겹살'))&(df['년도-월']=='2021-06')&(df['실판매규격'].str.contains('600g'))]
df_sam
```

```python
# 삼겹살 600g의 평균 가격은?
df_sam['가격(원)'].mean()
```

```python
# 삼겹살 600g의 최고 가격은?
df_sam['가격(원)'].max()
```

```python
df_sam[df_sam['가격(원)']<=5000]
```

```python
# 삼겹살 600g의 최저 가격은?
df_sam['가격(원)'].min()
```

## 우리동네 삼겹살 가격

```python
gu= input('구이름:')
```

```python
# 우리구 삼겹살 가격
df_sam_gu= df_sam[df_sam['자치구 이름']==gu][['시장/마트 이름','품목 이름','실판매규격','가격(원)']].drop_duplicates()
df_sam_gu
```

```python
# 시각화
x=df_sam_gu['시장/마트 이름']
y=df_sam_gu['가격(원)']

plt.title( gu+'시장별 삼겹살 가격 분포도')
plt.grid(True)
plt.scatter(x,y)
plt.show()
```

## 마트 지점별 삼겹살 가격

```python
mart= input('시장/마트 이름:')
```

```python
# 마트 지점별 삼겹살 가격
df_sam_mart=df_sam[df_sam['시장/마트 이름'].str.contains(mart)][['시장/마트 이름','품목 이름','실판매규격','가격(원)']].drop_duplicates()
df_sam_mart
```

```python
# 시각화
x=df_sam_mart['시장/마트 이름']
y=df_sam_mart['가격(원)']

plt.title('서울시'+mart+'별 삼겹살 가격분포도')
plt.xticks(rotation=30)
plt.grid(True)
plt.scatter(x,y)
plt.show()
```

# 달걀 가격 분석

## 달걀 데이터

```python
# 2021-06 데이터 이용
df_egg= df[(df['품목 이름'].str.contains('달걀'))& (df['년도-월']=='2021-06')
           & (df['실판매규격'].str.contains('30개'))& (df['가격(원)']>0)]
df_egg
```

```python
# 달걀 30구 평균 가격
df_egg['가격(원)'].mean()
```

```python
# 달걀 최고 가격
df_egg['가격(원)'].max()
```

```python
df_egg[df_egg['가격(원)']>20000]
```

```python
# 달걀 최저 가격
df_egg['가격(원)'].min()
```

## 우리동네 달걀 가격

```python
gu= input('자치구 이름:')
```

```python
# 우리구 달걀 가격
df_egg_gu= df_egg[df_egg['자치구 이름']==gu][['시장/마트 이름','품목 이름','실판매규격','가격(원)']].drop_duplicates()
df_egg_gu
```

```python
# 시각화
#df_egg_gu.sort_values('가격(원)')
x=df_egg_gu['시장/마트 이름']
y=df_egg_gu['가격(원)']

plt.figure(figsize=(8,5))
plt.title(gu+' 시장별 달걀 가격 분포도(30구)')
plt.scatter(x,y,color='y')
plt.grid(True)
plt.show()
```

```python
# 시각화
x=df_egg_mart['시장/마트 이름']
y=df_egg_mart['가격(원)']

plt.figure(figsize=(8,3))
plt.title('서울시 '+mart+'별 달걀 가격분포도(30구)')
plt.scatter(x,y,color='y')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
